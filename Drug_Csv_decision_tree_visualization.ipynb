```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

data = pd.read_csv("drug200.csv")

X = data[['Age']]
y = data['Drug']  

model = DecisionTreeClassifier(criterion="gini")
model.fit(X, y)

plt.figure(figsize=(12,8))
tree.plot_tree(model, feature_names=["Age"], class_names=model.classes_, filled=True)
plt.show()
```

```python
import pandas as pd  
from sklearn.tree import DecisionTreeClassifier  
from sklearn import tree  
import matplotlib.pyplot as plt  

# Load data  
data = pd.read_csv("drug200.csv")  

# Input features (Independent Variable)  
X = data[['Age']] # Selecting 'Age' as input feature  

# Target labels (Dependent Variable)  
y = data['Drug'] # Selecting 'Drug' as target label  

# Train a decision tree  
model = DecisionTreeClassifier(criterion="gini")  
model.fit(X, y)  

# Visualize the tree  
plt.figure(figsize=(12,8))  
tree.plot_tree(model, feature_names=["Age"], class_names=model.classes_, filled=True)  
plt.show()  
```
