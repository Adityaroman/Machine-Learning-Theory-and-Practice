```python
import pandas as pd
import seaborn as sns
data = {'Age':[22,25,29,24,23,27,30],
       'Salary':[4000,5000,000,73000,8000,8000,50000],
       'Experience':[1,3,5,7,6,9,4]}
df=pd.DataFrame(data)
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm") #correlation (columnwise)(corr shows +1,-1,0 values)positive means gradually incresse, negative is one increasing one decreasing,0 means no change no chamce to find relationship the values are like that
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Corrected Data
data = {
    'Age': [22, 25, 29, 24, 23, 27, 30],
    'Salary': [4000, 5000, 70000, 73000, 8000, 8000, 50000],  # Fixed salary
    'Experience': [1, 3, 5, 7, 6, 9, 4]
}

df = pd.DataFrame(data)

# Creating Subplots
fig, axs = plt.subplots(3, 3, figsize=(12, 10))  # 3x3 grid for 9 colormaps

# List of color maps
cmaps = ["Blues", "coolwarm", "BuPu", "Greens", "Oranges", "Reds", "Purples", "YlOrBr", "YlGnBu"]

# Generating heatmaps for each cmap
for ax, cmap in zip(axs.flat, cmaps):
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap=cmap, ax=ax)
    ax.set_title(f"cmap = {cmap}")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data (replace with your actual data)
data = {
    'Math': {'Male': 30, 'Female': 25},
    'Science': {'Male': 28, 'Female': 32},
    'History': {'Male': 22, 'Female': 28},
    'Computer': {'Male': 35, 'Female': 15},
    'English': {'Male': 20, 'Female': 30}
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
subjects = df.columns
male_counts = df.loc['Male']
female_counts = df.loc['Female']

width = 0.35  # Width of the bars

fig, ax = plt.subplots(figsize=(10, 6))  # Adjust figure size for better visualization

male_bars = ax.bar(subjects, male_counts, width, label='Male', color='skyblue')
female_bars = ax.bar(subjects, female_counts, width, bottom=male_counts, label='Female', color='lightcoral') #bottom = male count makes femal count on top of male count

# Add labels, title, and legend
ax.set_xlabel('Subject')
ax.set_ylabel('Number of Students')
ax.set_title('Number of Students by Subject and Gender')
ax.legend()

# Improve layout
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels if needed
plt.tight_layout() # Adjust layout to prevent labels from overlapping

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Math', 'Science', 'History', 'Computer', 'English']
male_counts = [22, 12, 22, 16, 25]
female_counts = [22, 12, 28, 16, 25]

plt.bar(categories, male_counts, label='Male', color='skyblue')
plt.bar(categories, female_counts, bottom=male_counts, label='Female', color='orange')

plt.xlabel('Subject')
plt.ylabel('Number of Students')
plt.title('Student Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

tight_layout() Explanation:

plt.tight_layout() is a very helpful function that automatically adjusts the plot's layout to prevent labels, titles, and other plot elements from overlapping the edges of the figure.  Here's why it's important:

```python
import matplotlib.pyplot as plt

categories = ['Math', 'Science', 'History', 'Computer', 'English']
male_counts = [22, 12, 22, 16, 25]
female_counts = [22, 12, 28, 16, 25]

plt.bar(categories, male_counts, label='Male', color='skyblue')
plt.bar(categories, female_counts, bottom=male_counts, label='Female', color='orange')

plt.xlabel('Subject')
plt.ylabel('Number of Students')
plt.title('Student Distribution')
plt.legend(loc='upper right')  # Place legend on the right
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Math', 'Science', 'History', 'Computer', 'English']
male_counts = [22, 12, 22, 16, 25]
female_counts = [22, 12, 28, 16, 25]

plt.bar(categories, male_counts, label='Male', color='skyblue')
plt.bar(categories, female_counts, bottom=male_counts, label='Female', color='orange')

plt.xlabel('Subject')
plt.ylabel('Number of Students')
plt.title('Student Distribution')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Place legend outside
plt.tight_layout()
plt.subplots_adjust(right=0.8)  # Adjust right margin
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]

# Define data for each line
lines = [
    {'y': [25, 30, 35, 40], 'color': 'green', 'marker': '.', 'linestyle': '-', 'label': 'Line 1'},
    {'y': [15, 25, 35, 45], 'color': 'blue', 'marker': 's', 'linestyle': '-', 'label': 'Line 2'},
    {'y': [20, 25, 30, 35], 'color': 'red', 'marker': 'o', 'linestyle': '-', 'label': 'Line 3'},
    {'y': [10, 20, 30, 40], 'color': 'purple', 'marker': '+', 'linestyle': '-', 'label': 'Line 4'},
]

for line in lines:
    plt.plot(x, line['y'],  # Use pre-calculated y values
             color=line['color'],
             linewidth=3,
             marker=line['marker'],
             markersize=15,
             linestyle=line['linestyle'],
             label=line['label'])

plt.title("Line Chart")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.legend()
plt.show()
```
