```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generate quadratic data with some noise
np.random.seed(0)
X = np.linspace(-3, 3, 20).reshape(-1, 1)  # Fixed typo: np.linespace -> np.linspace
y = X**2 + np.random.normal(0, 1, X.shape)

# Train two models: Linear (underfitting) and Polynomial (overfitting)
linear_model = LinearRegression()
poly_features = PolynomialFeatures(degree=5)
poly_model = LinearRegression()

# Fit the models
linear_model.fit(X, y)
X_poly = poly_features.fit_transform(X)
poly_model.fit(X_poly, y)

# Prediction
X_test = np.linspace(-3, 3, 100).reshape(-1, 1)  # Fixed typo: np.linespace -> np.linspace
y_linear_pred = linear_model.predict(X_test)
y_poly_pred = poly_model.predict(poly_features.transform(X_test))

# Plot results
plt.scatter(X, y, color='black', label='Actual Data')  # Fixed typo: X_y -> X, y
plt.plot(X_test, y_linear_pred, label="Underfitting (Linear)", color='blue')  # Fixed label wording
plt.plot(X_test, y_poly_pred, label="Overfitting (Polynomial Degree 5)", color='red')
plt.legend()  # Fixed typo: plt.legent() -> plt.legend()
plt.title("Underfitting vs Overfitting (Quadratic Data)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Set random seed for reproducibility
np.random.seed(0)

# Generate sample data
X = np.linspace(0, 10, 10).reshape(-1, 1)  # Fixed typo (linespace → linspace)
y = X**2 + np.random.normal(0, 5, size=X.shape)  # Add noise to simulate real data

# Fit models
lin_reg = LinearRegression().fit(X, y)  # Underfit (linear model)
poly2 = LinearRegression().fit(PolynomialFeatures(3).fit_transform(X), y)  # Good fit (quadratic)
poly8 = LinearRegression().fit(PolynomialFeatures(8).fit_transform(X), y)  # Overfit (high-degree)

# Predictions
y_pred_linear = lin_reg.predict(X)
y_pred_poly2 = poly2.predict(PolynomialFeatures(3).fit_transform(X))
y_pred_poly8 = poly8.predict(PolynomialFeatures(8).fit_transform(X))

# Plot results
plt.scatter(X, y, color='black', label="Data")  # Plot actual data points
plt.plot(X, y_pred_linear, label="Underfit (Linear)", color='blue')
plt.plot(X, y_pred_poly2, label="Good Fit (Quadratic)", color='green')
plt.plot(X, y_pred_poly8, label="Overfit (8-degree)", color='red')

# Add labels and legend
plt.legend()
plt.title("Underfitting vs Overfitting (Quadratic Data)")
plt.show()
```
