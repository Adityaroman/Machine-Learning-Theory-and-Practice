```python
import pandas as pd
a=pd.read_csv("C:/Users/adity_uvypdil/Data science/iris_data.csv")
a
```

```python
#Data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Bar chart
sns.barplot(data=a, x="species", y="petal_length")
plt.show()
```

```python
sns.barplot(data=a, x="species", y="petal_width")
plt.show()
```

```python
#pi Chart
species_counts = a['species'].value_counts()
print(species_counts)

#print(species_counts.index)
#plt.pie(species_counts, labels=species_counts.index, autopct='%.2f%%')
#plt.show()
```

```python
species_counts = a['species'].value_counts()
print(species_counts)
print(species_counts.index)
plt.pie(species_counts, labels=species_counts.index, autopct='%.2f%%')
plt.show()
```

```python
#scatter plot
sns.scatterplot(data=a,x='petal_length', y='petal_width', hue='species')
plt.show()
```

```python
#Histogram
plt.hist(a['species'],color='skyblue',edgecolor='black')
plt.show()
```
