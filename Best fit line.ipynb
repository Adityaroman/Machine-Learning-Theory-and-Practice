**1. Set up the data**

* x values: 1, 2, 3, 4, 5
* y values: 3, 4, 2, 4, 5

**2. Calculate the means**

* Mean of x (x̄) = (1 + 2 + 3 + 4 + 5) / 5 = 3
* Mean of y (ȳ) = (3 + 4 + 2 + 4 + 5) / 5 = 3.6

**3. Calculate the slope (m)**

The formula for the slope is:

```
m = Σ[(xi - x̄)(yi - ȳ)] / Σ[(xi - x̄)²] 
```

Let's break that down:

* **(xi - x̄)**:  For each x value, subtract the mean of x.
    * 1 - 3 = -2
    * 2 - 3 = -1
    * 3 - 3 = 0
    * 4 - 3 = 1
    * 5 - 3 = 2
* **(yi - ȳ)**: For each y value, subtract the mean of y.
    * 3 - 3.6 = -0.6
    * 4 - 3.6 = 0.4
    * 2 - 3.6 = -1.6
    * 4 - 3.6 = 0.4
    * 5 - 3.6 = 1.4
* **[(xi - x̄)(yi - ȳ)]:** Multiply the corresponding results from above.
    * (-2) * (-0.6) = 1.2
    * (-1) * (0.4) = -0.4
    * (0) * (-1.6) = 0
    * (1) * (0.4) = 0.4
    * (2) * (1.4) = 2.8
* **Σ[(xi - x̄)(yi - ȳ)]:** Sum the results from the previous step.
    * 1.2 + (-0.4) + 0 + 0.4 + 2.8 = 4
* **[(xi - x̄)²]:** Square each of the (xi - x̄) values.
    * (-2)² = 4
    * (-1)² = 1
    * (0)² = 0
    * (1)² = 1
    * (2)² = 4
* **Σ[(xi - x̄)²]:** Sum the squared values.
    * 4 + 1 + 0 + 1 + 4 = 10

Now, plug the sums into the formula for the slope (m):

```
m = 4 / 10 = 0.4
```

**4. Calculate the y-intercept (c)**

The formula for the y-intercept is:

```
c = ȳ - m * x̄
```

Substitute the values we know:

```
c = 3.6 - 0.4 * 3 = 2.4
```

**5. The best-fit line**

Now we have the equation of our best-fit line:

```
y = mx + c 
y = 0.4x + 2.4
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 4, 5])

# Calculate means
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate slope (m)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)
m = numerator / denominator

# Calculate y-intercept (c)
c = y_mean - m * x_mean

# Best-fit line equation
def best_fit_line(x_val):
    return m * x_val + c

# Generate points for the line for plotting
x_line = np.linspace(x.min(), x.max(), 100)  # Creates evenly spaced points
y_line = best_fit_line(x_line)

# Plot the data and the best-fit line
plt.scatter(x, y, label="Data Points")
plt.plot(x_line, y_line, color="red", label="Best-Fit Line")

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)  # Add a grid for better readability

# Show the plot
plt.show()

# Print the equation of the line
print(f"The best-fit line is: y = {m:.2f}x + {c:.2f}")


#Demonstration of using the best_fit_line function
new_x = 2.5
predicted_y = best_fit_line(new_x)
print(f"For x = {new_x}, the predicted y is: {predicted_y:.2f}")
````
