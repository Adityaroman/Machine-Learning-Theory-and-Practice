```python
import matplotlib.pyplot as plt  #for seaborn import seaborn as sns
```

```python
#Line chart
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]
plt.plot(x,y)
plt.title("Line Chart")
plt.ylabel('Y-Axis')
plt.xlabel('x-Axis')
plt.show()
```

```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]
plt.plot(x,y, color='green', linewidth=3, marker= '*', markersize=15, linestyle= '--')
plt.title("Line Chart")
plt.ylabel('Y-Axis')
plt.xlabel('x-Axis')
plt.show()
```

```python
#BAr Chart
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]
plt.bar(x,y)
plt.title("Bar Chart")
plt.ylabel('Y-Axis')
plt.xlabel('x-Axis')
plt.show()
```

```python
#
L = ['a','b','c','d','e',]
data = [20, 25, 35, 55, 44]
plt.pie(data, labels=L)
plt.title("pie Chart")
plt.show()
```
